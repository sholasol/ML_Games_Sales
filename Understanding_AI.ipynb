### Dealing with Missing Values

```python
# Handling the missing name
# Drop columns with missing names
new_df = game_df.dropna(subset=['Name'])
```

## Handling Publisher
#### Drop column with missing Publisher

```python
drop_publisher = new_data.dropna(subset=['Publisher'])
```

```python
drop_publisher.isna().sum()
```

## Handling Critic Score

```python
# Fill critic score column with missing value with median critic score

critic_score_median = drop_publisher['Critic_Score'].median()

drop_publisher['Critic_Score'] = drop_publisher['Critic_Score'].fillna(critic_score_median)
```

## Handling Critic_Count

```python
# Fill Critic_Count column with missing value with median critic count

critic_count_median = drop_publisher['Critic_Count'].median()

drop_publisher['Critic_Count'] = drop_publisher['Critic_Count'].fillna(critic_count_median)
```

## Handling User_Score

```python
# Fill User Score column with missing value with median user score

# The user score contains a record 'tbd' of string type, the values has to be converted to NaN, then all NaN record 
# are filled with median of the score

drop_publisher['User_Score'] = drop_publisher['User_Score'].replace('tbd', np.nan)

drop_publisher['User_Score'] = drop_publisher['User_Score'].astype(float)

user_score_median = drop_publisher['User_Score'].median()

drop_publisher['User_Score'] = drop_publisher['User_Score'].fillna(user_score_median)
```

## Handling User Count

```python
# Fill User Count column with missing value with median user count

user_count_median = drop_publisher['User_Count'].median()

drop_publisher['User_Count'] = drop_publisher['User_Count'].fillna(user_count_median)
```

```python
#### Regression Model(Creating feature matrix X and target) vector y
# Select the features to be used in the regression model as feature Matrix X

X = game_record_numeric[['Critic_Score', 'User_Score', 'Year_of_Release', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

y = game_record_numeric['Global_Sales']

scaler = StandardScaler() # Creating instance of the StandardScaler() class

X_scale = scaler.fit_transform(X)

# Split the model into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X_scale,y,test_size=0.2, random_state=42)
```

```python
label_encoder = LabelEncoder()
for column in game_record.select_dtypes(include='object').columns:
    game_record[column] = label_encoder.fit_transform(game_record[column])

# Define the target variables
targets = ['Platform', 'Genre', 'Publisher', 'Developer', 'Rating']

# Train a logistic regression model for each target variable
for target in targets:
    # Split the data into training and testing sets
    X_train = game_record.drop(columns=[target, 'Name', 'Year_of_Release'])
    y_train = game_record[target]
    X_test = game_record.drop(columns=[target, 'Name', 'Year_of_Release']).sample(frac=0.2, random_state=1)
    y_test = game_record[target].loc[X_test.index]

    # Scale the predictors using StandardScaler
    #scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Train the model
    logisticReg = LogisticRegression(max_iter=1000)
    logisticReg.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = logisticReg.predict(X_test)

    # Evaluate the logisticReg's performance
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')

    # Print the results
    print(f'Target Variable: {target}')
    print(f'Accuracy: {accuracy:.2f}')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print('\n')
```

```python
# Define the target variables
targets = ['Platform', 'Genre', 'Publisher', 'Developer', 'Rating']

# Create empty lists to store the performance metrics
train_scores = []
test_scores = []

# Train a logistic regression model for each target variable
for target in targets:
    # Split the data into training and testing sets
    X_train = game_record.drop(columns=[target, 'Name', 'Year_of_Release'])
    y_train = game_record[target]

    # Scale the predictors using StandardScaler
    #scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)

    # Train the model
    logisticReg = LogisticRegression(max_iter=1000)

    # Use cross-validation to compute the performance metrics
    train_score = cross_val_score(logisticReg, X_train, y_train, cv=5, scoring='accuracy')
    test_score = cross_val_score(logisticReg, X_train, y_train, cv=5, scoring='accuracy')

    # Store the performance metrics in the lists
    train_scores.append(train_score.mean())
    test_scores.append(test_score.mean())

    # Print the results
    print(f'Target Variable: {target}')
    print(f'Training Scores: {train_score}')
    print(f'Testing Scores: {test_score}')
    print(f'Mean Training Score: {train_scores[-1]:.2f}')
    print(f'Mean Testing Score: {test_scores[-1]:.2f}')
    print('\n')
```

### Internal evaluation metrics to determine which categorical variable best describes the groups formed.
#### The use of Silhouette Coefficient and David-Bouldin Index Using different Number of Clustering (n_clusters = [3, 4, 5])

#### Categorical Variable - Platform          
#### Non-Categorical - Critic Score

```python
# Encode the categorical variable "Platform"
le = LabelEncoder()
game_record['Platform_Encoded'] = le.fit_transform(game_record['Platform'])

# Define the variables to be used in clustering
variables = ['Critic_Score', 'Platform_Encoded']

# Standardize the data
scaled_data = (game_record[variables] - game_record[variables].mean()) / game_record[variables].std()

# Perform K-Means clustering with 3 clusters
k3 = KMeans(n_clusters=3, random_state=42)
k3.fit(scaled_data)
game_record['Cluster_3'] = k3.labels_

# Perform K-Means clustering with 4 clusters
k4 = KMeans(n_clusters=4, random_state=42)
k4.fit(scaled_data)
game_record['Cluster_4'] = k4.labels_

# Perform K-Means clustering with 5 clusters
k5 = KMeans(n_clusters=5, random_state=42)
k5.fit(scaled_data)
game_record['Cluster_5'] = k5.labels_

# Evaluate the clustering using the silhouette score
silhouette_3 = silhouette_score(scaled_data, k3.labels_)
silhouette_4 = silhouette_score(scaled_data, k4.labels_)
silhouette_5 = silhouette_score(scaled_data, k5.labels_)

# Evaluate the clustering using the Davies-Bouldin index
db_3 = davies_bouldin_score(scaled_data, k3.labels_)
db_4 = davies_bouldin_score(scaled_data, k4.labels_)
db_5 = davies_bouldin_score(scaled_data, k5.labels_)

# Print the results
print("Silhouette Score for K-means3: {}".format(silhouette_3))
print("Davies-Bouldin Score for K-means3: {}".format(db_3))
print('-------------------------------------------------------')
print("Silhouette Score for K-means4: {}".format(silhouette_4))
print("Davies-Bouldin Score for K-means4: {}".format(db_4))
print('-------------------------------------------------------')
print("Silhouette Score for K-means5: {}".format(silhouette_5))
print("Davies-Bouldin Score for K-means5: {}".format(db_5))
```
